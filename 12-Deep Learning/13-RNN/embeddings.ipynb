```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sentences = [
    'the glass of milk',
    'the glass of juice',
    'the cup of tea',
    'I am a good boy',
    'I am a good developer',
    'understand the meaning of words',
    'your videos are good'
]
```

```python
## Define the vocab size
vocab_size = 10000
```

```python
##One hot representation for every word
one_hot_rep = [one_hot(words, vocab_size) for words in sentences]
one_hot_rep
```

```python
#Word Enbedding representation
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
sent_length = 8
embedded = pad_sequences(one_hot_rep, padding='pre', maxlen=sent_length)
print(embedded)
```

```python
##Feature Representation
dim = 10
model = Sequential()
model.add(Embedding(vocab_size,dim,input_length=sent_length))
model.compile(optimizer='adam',loss='mse')
```

```python
model.summary()
```

```python
model.predict(embedded)
```
