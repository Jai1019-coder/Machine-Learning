```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## creating dataset
from sklearn.datasets import make_classification
X,Y = make_classification(n_samples = 1000, n_classes=2)
```

```python
##train test split 
from sklearn.model_selection import train_test_split
X_train,x_test,Y_train,y_test = train_test_split(X,Y,test_size=0.25)
```

```python
## create a dummy model with 0 as output for any test imput
dummy_model_poba = [0 for _ in range(len(x_test))]
dummy_model_poba
```

```python
# logistic regression model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,Y_train)
```

```python
model_proba = model.predict_proba(x_test)
model_proba
```

```python
## We need to focus on probability corresponding to class 1
model_proba_1 = model_proba[:,1]
model_proba_1
```

```python
## getting scores 
from sklearn.metrics import roc_auc_score
dummy_score = roc_auc_score(y_test,dummy_model_poba)
print(dummy_score)
model_score = roc_auc_score(y_test, model_proba_1)
print(model_score)
```

```python
# plotting roc_curve
from sklearn.metrics import roc_curve
dummy_fpr, dummy_tpr,dummy_thresholds = roc_curve(y_test,dummy_model_poba)
print(dummy_fpr)
print(dummy_tpr)
print(dummy_thresholds)
```

```python
model_fpr, model_tpr,model_thresholds = roc_curve(y_test,model_proba_1)
print(model_fpr)
print(model_tpr)
print(model_thresholds)
```

```python
# import seaborn as sns
```

```python
plt.plot(dummy_fpr,dummy_tpr,linestyle = '--',label='Dummy Model')
plt.plot(model_fpr,model_tpr,marker='*',label='LogisticRegression Model')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.show()
```

```python
##plotting with annotations
fig = plt.figure(figsize=(20,50))
ax = fig.add_subplot(111)
plt.plot(dummy_fpr,dummy_tpr,linestyle = '--',label='Dummy Model')
plt.plot(model_fpr,model_tpr,marker='*',label='LogisticRegression Model',linewidth=4)
for xyz in zip(model_fpr,model_tpr,model_thresholds):
    ax.annotate('%s' % np.round(xyz[2],3),xy=(xyz[0],xyz[1]))
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.show()
```
